```python
X_train = X_train.fillna(X_train.mean())
```

```python
y_train = y_train.fillna(y_train.median())
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
X_test = X_test.to_frame()
```

```python
X_test = X_test.fillna(X_test.mean())
```

```python
y_test = y_test.fillna(y_test.median())
```

```python
y_pred = model.predict(X_test)
```

```python
print("Mean Absolute Error (MAE):", mean_absolute_error(y_test, y_pred))
print("Mean Squared Error (MSE):", mean_squared_error(y_test, y_pred))
print("Root Mean Squared Error (RMSE):", np.sqrt(mean_squared_error(y_test, y_pred)))
print("R² Score:", r2_score(y_test, y_pred))
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x=y_test, y=y_pred, color='blue')
plt.plot(y_test, y_test, color='red', linestyle='dashed')  # Ideal fit line
plt.xlabel("Actual Electric Range")
plt.ylabel("Predicted Electric Range")
plt.title("Actual vs. Predicted Electric Range")
plt.show()
```

> 0.75-Strong predictive power
0.50 - 0.75	Moderate predictive power
< 0.50-Weak predictive power (Model is not capturing patterns well)
< 0-Model is worse than predicting the mean (Re-evaluate features or data quality)
